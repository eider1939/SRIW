```python
from rdflib import Graph, Namespace

g = Graph()
```

```python
from rdflib import RDFS, XSD

g.bind("rdfs", RDFS)
g.bind("xsd", XSD)

g.bind("wdt", "http://www.wikidata.org/prop/direct/")
g.bind("wd", "http://www.wikidata.org/entity/")
g.bind("wikibase", "http://wikiba.se/ontology#")
g.bind("bd", "http://www.bigdata.com/rdf#")

g.bind("dbo", "http://dbpedia.org/ontology/")
g.bind("dbr", "http://dbpedia.org/resource/")
g.bind("dbp", "http://dbpedia.org/property/")
g.bind("dbd", "http://dbpedia.org/datatype/")
```

```python
#cosulta a wikidata con fin de obtencion de imagenes 

from rdflib import Graph, Namespace

#sentencia de consulta Sparql
q1 = """      
SELECT  ?motoLabel ?img ?imgLabel ?manufactureLabel 
WHERE
{
    SERVICE <https://query.wikidata.org/sparql> {
        ?moto wdt:P31 wd:Q23866334.
        ?moto wdt:P18 ?img.
        ?moto wdt:P176 ?manufacturer.
        ?moto rdfs:label ?motoLabel.
        ?manufacturer rdfs:label ?manufactureLabel.
        FILTER(LANG(?motoLabel) = 'en' ).
        FILTER(LANG(?manufactureLabel) = 'en' ).
    }
} 
"""

#consulta en el grafo
res1 = g.query(q1)
#imprecion de los resultados
lista_motos=[]  #almacena las url de las imagenes
for row in res1:
    l=[]
    #print(f"Item: {row.motoLabel}, Img: {row.img}, Manufacture: {row.manufactureLabel}")
    l.append(row.motoLabel) #agregamos el nombre de la moto
    l.append(row.img)   #agregamos al imagenes a la l lista 
    lista_motos.append(l)
```

```python
import urllib.request

ruta="/content/drive/MyDrive/Redes Neuronales/Datos/Vehiculos/Motos/"
cont_fallos=0     #contador de fallos
for i in range(0,len(lista_motos)):
  ruta1=ruta+lista_motos[i][0]+'.jpg'
  try:
    f = open(ruta1,'wb')
    f.write(urllib.request.urlopen(lista_motos[i][1]).read())
    f.close()
  except:
    cont_fallos+=1


```
